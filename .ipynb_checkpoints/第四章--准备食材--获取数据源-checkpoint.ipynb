```python
df4=pd.read_excel(r".\file\大学排名.xlsx",sheet_name=0,header=0) # 使用第一行作为列索引
df4
```

```python
df4=pd.read_excel(r".\file\大学排名.xlsx",sheet_name=0,header=1) # 使用第2行作为列索引
df4
```

```python
df4=pd.read_excel(r".\file\大学排名.xlsx",sheet_name=0,header=None) # 默认从0开始作为列索引
df4
```

```python
df5=pd.read_excel(r".\file\大学排名.xlsx",usecols=1)
df5
```

```python
df6=pd.read_excel(r".\file\大学排名.xlsx",usecols=[0,2])  # 导入 第 0列和第1列
df6
```
