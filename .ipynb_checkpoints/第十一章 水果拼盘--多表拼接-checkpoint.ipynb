##### 连接键的类型

```python
df1=pd.read_table(r".\file\test_111.txt",sep=" ",encoding="utf-8")
df=pd.read_table(r".\file\test_114.txt",sep=" ",encoding="utf-8")
pd.merge(df,df1,left_on="编号",right_on="学号")   # 指定左右连接键
```

```python
pd.merge(df,df1,left_index=True,right_index= True)  # 按索引列当做连接键
```

```python
# pd.merge(df,df1,left_index=True,right_on="学号")   也可以混用
```

##### 连接方式

```python
df1=pd.read_table(r".\file\test_1115.txt",sep=" ",encoding="utf-8")   # 内连接
df=pd.read_table(r".\file\test_112.txt",sep=" ",encoding="utf-8")
pd.merge(df,df1,on="学号",how="inner")
```

```python
pd.merge(df,df1,on="学号",how="left")  # 左连接
```

```python
pd.merge(df,df1,on="学号",how="outer")  # 外连接
```

#### 表的纵向拼接

```python
pd.concat([df,df1])  # 普通合并
```

```python
pd.concat([df,df1],ignore_index=True) # 索引设置
```

```python
pd.concat([df,df1],ignore_index=True).drop_duplicates()  # 重叠数据合并，下面没有重叠的，所以看不出来效果。懒的在去编数据了
```
