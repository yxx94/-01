#### 时间索引

```python
import pandas as pd
import numpy as np
index = pd.DatetimeIndex(["2020-01-01",'2020-01-02','2020-01-05','2019-05-08','2019-12-30','2019-12-25'])
data=pd.DataFrame(np.arange(1,7),columns=['num'],index=index)
data
```

```python
data['2019'] # 获取2019年的数据
```

```python
data['2019-12'] # 获取2019年12月的数据
```

```python
# 获取2019年12月的数据到2020年1月的数据
data['2019-12':'2020-01']
```

##### 时间偏移

```python
from datetime import timedelta
data = datetime(2018,5,18,20,32)
data + timedelta(days=1)  # 往后推一天
```

```python
data + timedelta(seconds=60) # 往后推60s
```

```python
# 往前推一天
data-timedelta(days=1)
```

```python
from pandas.tseries.offsets import Day,Hour,Minute
date=datetime(2018,5,18,20,32)
```

```python
date + Day(1) # 往后推1天
```

```python
date + Hour(1) # 往后推一个小时
```

```python
date - Day(1) # 往前推一天
```

```python
date - Minute(10) # 往前推 10分钟
```
